```python
import dill

import matplotlib.pyplot as plt

from src import config as cfg
```

```python
model_names = [
    'lgb_1',
    'xgb_1',
#     'rf_1',
]

models = {}

for name in model_names:
    with open(f'{cfg.models_path}/model_{name}_last.pkl', 'rb') as f:
        models[name] = dill.load(f)
    print(name, models[name].log['test_score'].min())
```

```python
for name, model in models.items():
    print(name, model.log.time.sum())
    display(model.log.sort_values('test_score', ascending=True).drop(columns=['hyperparam_dict', 'timestamp']).head(50))
```
